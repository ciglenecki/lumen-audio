```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
convnext_small = json.loads(open("../../data/convnext_small_infer_res.json").read())
resnext = json.loads(open("../../data/resnext_infer_res.json").read())
effnet = json.loads(open("../../data/effnet_s_infer_res.json").read())
mobnet = json.loads(open("../../data/mobnet_infer_res.json").read())
```

```python
def plot_means_from_json(json_file, label):
    steps = np.array([i["step"] for i in json_file])
    means = [i["mean"] for i in json_file]
    stds = [i["std"] for i in json_file]
    
    plt.plot((steps+1)*10, means, label=label)

plt.figure(figsize=(11, 5))

plot_means_from_json(convnext_small, "ConvNeXt small")
plot_means_from_json(resnext, "ResNeXt50")
plot_means_from_json(effnet, "EfficientNet V2 small")
plot_means_from_json(mobnet, "MobileNet V3 large")

plt.title("The time required for forward pass compared to the length of the audio sequence")
plt.xlabel("Length of the audio (sec)")
plt.ylabel("Mean time required for forward pass (sec)")
plt.legend()
plt.show()
```

```python
def plot_std_from_json(json_file, label):
    steps = np.array([i["step"] for i in json_file])
    means = [i["mean"] for i in json_file]
    stds = [i["std"] for i in json_file]
    
    plt.plot((steps+1)*10, stds, label=label)

    
plt.figure(figsize=(11, 5))

plot_std_from_json(convnext_small, "ConvNeXt small")
plot_std_from_json(resnext, "ResNeXt50")
plot_std_from_json(effnet, "EfficientNet V2 small")
plot_std_from_json(mobnet, "MobileNet V3 large")

plt.xlabel("Length of the audio (sec)")
plt.ylabel("Stand. dev. of time required for forward pass (sec)")
plt.legend()
plt.show()
```
